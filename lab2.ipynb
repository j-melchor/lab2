#### **Section 4: Evaluate the model**

*Objective:* Evaluate the model with the fitted data.

Task: build the Predict function, and Calculate accuracy.

```python
# Predict function
def predict (X, theta):
  X_bias = np.column_stack((np.ones(X.shape[0]), X))
  probabilities = 1 / (1 + np.exp(-X_bias.dot(theta)))
  return (probabilities >= 0.5).astype(int)

# Make predictions
theta_optimized = np.hstack((model.intercept_, model.coef_.flatten()))
predictions = predict(X_test, theta_optimized)
```

```python
# Calculate accuracy
accuracy = np.mean(predictions == y_test) * 100
print("Accuracy:", accuracy, "%")
```
